1. What is Python, and why is it popular?
Python is a high-level, interpreted programming language known for its readability, simplicity, and versatility. It supports multiple programming paradigms, including procedural, object-oriented, and functional programming. Python is popular due to its user-friendly syntax, large community, and vast collection of libraries, making it suitable for everything from web development to data science and artificial intelligence.

2. What is an interpreter in Python?
An interpreter in Python is a program that reads and executes Python code line by line. Unlike compilers, which translate the entire code at once, interpreters execute the code directly, making Python an interpreted language. This allows for quick testing and debugging, but it can be slower than compiled languages.

3. What are pre-defined keywords in Python?
Pre-defined keywords in Python are reserved words that have a special meaning and cannot be used as variable names. These keywords define the syntax and structure of the Python language. Examples include if, else, def, while, for, class, try, and more.

4. Can keywords be used as variable names?
No, keywords cannot be used as variable names because they are reserved by the Python language for specific purposes. Attempting to use a keyword as a variable name will result in a syntax error.

5. What is mutability in Python?
Mutability refers to whether the state or content of an object can be changed after it is created. Immutable objects cannot be modified once created, while mutable objects can be changed. In Python, mutable objects include lists, dictionaries, and sets, while immutable objects include integers, strings, and tuples.

6. Why are lists mutable, but tuples are immutable?
Lists are mutable in Python because they are designed to allow dynamic changes, such as adding, removing, or changing elements. Tuples, on the other hand, are immutable by design. This immutability helps improve performance and ensures that tuples can be used as keys in dictionaries or elements in sets, where hashable (immutable) objects are required.

8. What are logical operators in Python?
Logical operators in Python are used to combine conditional statements and perform logical operations:

and: Returns True if both conditions are true.
or: Returns True if at least one condition is true.
not: Reverses the result of a condition (True becomes False, False becomes True).

11. What is the purpose of conditional statements in Python?
Conditional statements, such as if, elif, and else, are used to control the flow of a program. They allow the program to execute different code based on certain conditions.

 Write a program that checks if a given word is a Python keyword

```python
import keyword

def is_python_keyword(word):
    return keyword.iskeyword(word)

def main():
    word = input("Enter a word to check if it is a Python keyword: ").strip()
    if is_python_keyword(word):
        print(f"'{word}' is a Python keyword.")
    else:
        print(f"'{word}' is NOT a Python keyword.")

if __name__ == "__main__":
    main()
```

Write a Python program to convert user input from string to integer, float, and boolean types0


```python
# Get user input
user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print(f"Integer value: {int_value}")
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print(f"Float value: {float_value}")
except ValueError:
    print("Cannot convert to float.")

# Convert to boolean
bool_value = user_input.lower() in ['true', '1', 'yes', 'y', 'on']
print(f"Boolean value: {bool_value}")
```

 Write code to demonstrate type casting with list elements

```python
# Type casting with list elements
input_list = input("Enter a list of values separated by space: ").split()

# Convert list elements to integer, float, and boolean
def convert_list_elements(lst):
    int_list = []
    float_list = []
    bool_list = []

    for item in lst:
        try:
            int_list.append(int(item))
        except ValueError:
            int_list.append(None)

        try:
            float_list.append(float(item))
        except ValueError:
            float_list.append(None)

        bool_list.append(item.lower() in ['true', '1', 'yes', 'y', 'on'])

    return int_list, float_list, bool_list

int_values, float_values, bool_values = convert_list_elements(input_list)

print(f"Integer List: {int_values}")
print(f"Float List: {float_values}")
print(f"Boolean List: {bool_values}")
```

Write a program that checks if a number is positive, negative, or zero

```python
def check_number_sign(number):
    try:
        num = float(number)
        if num > 0:
            print("The number is positive.")
        elif num < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

num_input = input("Enter a number to check its sign: ")
check_number_sign(num_input)
```

 Write a program to reverse a string using a while loop0

```python
def reverse_string(s):
    reversed_str = ""
    index = len(s) - 1
    while index >= 0:
        reversed_str += s[index]
        index -= 1
    return reversed_str

string_input = input("Enter a string to reverse: ")
print(f"Reversed string: {reverse_string(string_input)}")
```

Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    result = 1
    while n > 1:
        result *= n
        n -= 1
    return result

num = int(input("Enter a number to calculate its factorial: "))
print(f"Factorial of {num} is {factorial(num)}")
```
